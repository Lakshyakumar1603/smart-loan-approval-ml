Steap-5 : Feature Engineering and Preprocessing 

```python
df.info()
```

```python
df['debt_to_income'] = round(df[' loan_amount'] / df[' income_annum'],5)
```

```python
df
```

```python
def cibil_category(score):
    if score < 600: return 'Low'
    elif score < 750: return 'Medium'
    else: return 'High'
#apply function are used 
df['cibil_cat'] = df[' cibil_score'].apply(cibil_category)
```

```python
df
```

```python
#outlier
Q1 = df[' loan_amount'].quantile(0.25)
Q3 = df[' loan_amount'].quantile(0.75)
IQR = Q3 - Q1
df = df[(df[' loan_amount'] >= Q1 - 1.5*IQR) & (df[' loan_amount'] <= Q3 + 1.5*IQR)]
```

```python
df
```

Steap-8 : Model Selection

```python
#classification algorithims are applied 
from sklearn.linear_model import  LinearRegression
from sklearn.ensemble import RandomForestRegressor,AdaBoostRegressor,GradientBoostingRegressor
from sklearn.tree import DecisionTreeRegressor
from xgboost import XGBRegressor
```

Steap-9 : Model Training 

```python
#create the object each of them 
lnr=LinearRegression()
rfc=RandomForestRegressor()
dtc=DecisionTreeRegressor()
abc=AdaBoostRegressor()
gbc=GradientBoostingRegressor()
xgc=XGBRegressor()
```

```python
l=[lnr,rfc,dtc,abc,gbc,xgc]
```

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
for model in l:
    print(f"\n🛠 Training Model: {model.__class__.__name__}")
    model.fit(x_train, y_train)
    print(f"Training Score of {model.__class__.__name__} is {model.score(x_train, y_train)}")
    print(f"Testing Score of {model.__class__.__name__} is {model.score(x_test, y_test)}")
    
```

```python
import pandas as pd

# Create an empty list to store results
results = []

# Train each model and collect results
for model in l:
    model.fit(x_train, y_train)
    train_score = model.score(x_train, y_train)
    test_score = model.score(x_test, y_test)
    
    results.append({
        'Model': model.__class__.__name__,
        'Training Score': round(train_score, 4)*100,
        'Testing Score': round(test_score, 4)*100
    })

# Convert the results into a DataFrame
results_df = pd.DataFrame(results)

# Display the DataFrame
print("\n📊 Model Evaluation Summary:")
print(results_df)
```
